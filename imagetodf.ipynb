```python
# importando as bibliotecas necessárias
import cv2
import pytesseract
import pandas as pd

# Definindo o caminho para o executável do Tesseract OCR
caminho = r"C:\Program Files\Tesseract-OCR"
pytesseract.pytesseract.tesseract_cmd = caminho + r"\tesseract.exe"

# Lendo a imagem
imagem = cv2.imread("rrd_ipoa2.png")

# Verificando se a imagem foi carregada com sucesso
if imagem is not None:
    # Extraindo o texto da imagem
    texto = pytesseract.image_to_string(imagem)
    
    # Dividindo o texto em linhas
    linhas = texto.split('\n')
    
    # Criando uma lista para armazenar os dados da tabela
    dados = []
    
    # Processando cada linha e dividir em colunas
    for linha in linhas:
        colunas = linha.split()
        if len(colunas) >= 4:  # Certificando de que há pelo menos 4 colunas de dados
            dados.append(colunas)
    
    # Convertendo a lista de dados em um DataFrame pandas
    df = pd.DataFrame(dados, columns=["Year", "Conservative", "Total_yearly_area", "Total_cumulative_area", "Column5", "Column6", "Column7"])
    
```

```python
# Para apagar as colunas 5, 6 e 7 (índices 4, 5 e 6) de forma inplace
df.drop(columns=["Column5", "Column6", "Column7"], inplace=True)
df.head()
```

```python
# Remove a primeira linha com strings
df = df.drop(0)
```

```python
# Remove vírgulas à direita na coluna "Total_cumulative_area"
df["Total_cumulative_area"] = df["Total_cumulative_area"].str.rstrip(',')
```

```python
# Remove o caractere ‘ (aspas curvas) da coluna "Total_cumulative_area" e converte para float
df["Total_cumulative_area"] = df["Total_cumulative_area"].str.replace('‘', '').astype(float)
```
